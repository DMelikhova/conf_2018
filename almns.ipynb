```python
from Bio.Blast import NCBIXML
```

```python
result_handle = open('W97DCBHH01R-Alignment.xml')
```

```python
blast_records = NCBIXML.parse(result_handle)
```

```python
good_almns = {}
for blast_record in blast_records:
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            print(hsp.align_length)
            if 62 >= hsp.query_start and(hsp.query[62 - hsp.query_start] != hsp.sbjct[62 - hsp.query_start]):
                info = [hsp.query_start, hsp.sbjct]
                good_almns[alignment.title] = info
```

```python
good_almns_new = {}
for k, v in good_almns.items():
    if 59 >= v[0] and len(v[1]) >= 123 - v[0] and v[1][59 -  v[0]] == 'G' and v[1][61 -  v[0]] == 'N' and v[1][63 -  v[0]] == 'R' and v[1][64 -  v[0]] == 'F' and \
       v[1][65 -  v[0]] == 'L' and v[1][66 -  v[0]] == 'Q' :
        good_almns_new[k] = v
```
